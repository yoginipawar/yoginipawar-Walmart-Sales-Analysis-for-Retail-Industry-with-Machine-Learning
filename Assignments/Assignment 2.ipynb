# **Find the Outliers and replace them with Outliers**

```python
columns_to_check = ['culmen_length_mm', 'culmen_depth_mm']
```

```python
def detect_outliers(data, column):
    Q1 = np.percentile(data[column], 25)
    Q3 = np.percentile(data[column], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

```python
outliers_dict = {}
for column in columns_to_check:
    outliers_dict[column] = detect_outliers(data, column)
```

```python
for column, outliers in outliers_dict.items():
  f"Outliers in {column}:"
  outliers
  "\n"
```

```python
def display_outliers(data, columns):
    for column in columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=data[column], orient='h', width=0.5)
        plt.title(f'Boxplot of {column}')
        plt.xlabel(column)
        plt.show()
```

```python
display_outliers(data, columns_to_check)
```
